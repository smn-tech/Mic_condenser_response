```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm
from matplotlib.ticker import LinearLocator
print('Libraries imported ...')
```

```python
# Number of simulation points:
N=1000
R1=1e3
Cd=10e-6
R2=1e3

# Mic capaciatnce:
C=np.linspace(5e-12,10e-12,N)
# Frequency:
F=np.linspace(20,8000,N)

F,Cm=np.meshgrid(F,C)
W=2*np.pi*F

img_part=W*R1*(Cm+Cd)
real_part=np.ones(img_part.shape)
A=real_part+(img_part*1j)

img_part=W*Cd
real_part=(W**2)*R1*Cd*Cm
B=R2*(real_part-(img_part*1j))

den_part=1-(A/B)
den_mag=np.abs(den_part)

G=1/den_mag


# Plotting section:
# fig=plt.figure(figsize=plt.figaspect(0.1))
# fig=plt.figure(figsize=(8,5))
# ax=fig.add_subplot(1,1,1,projection='3d')

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,7))


# Plot the surface.
surf = ax.plot_surface(F/1000, Cm/1e-12, G, cmap=cm.winter,
                       linewidth=0, antialiased=False)
# fig.colorbar(surf, shrink=0.5, aspect=20)


ax.set_xlabel('F(kHz)',fontsize=14, labelpad=5)
ax.set_ylabel('$C_m$ (pF)',fontsize=14, labelpad=5)
ax.set_zlabel('Gain',fontsize=14, labelpad=10, rotation=90)
ax.tick_params(axis='both', which='major', labelsize=12)

file_name='Gain plot for $R_1$='+str(R1)+' Ohm, $R_2$='+str(R2)+' Ohm'
ax.set_title(file_name,fontsize=14)
plt.show()
```
